```python
import cv2
from matplotlib import pyplot as plt 
import easyocr
import numpy as np
 
```

# **Fotoğrafı Okuma**

```python
IMAGE_PATH='surf.jpeg'
```

```python
reader=easyocr.Reader(['en'],gpu=False)
result=reader.readtext(IMAGE_PATH)
result #en sondaki değer confidence ilk kısımdakiler ise yazı fotoğrafın neresinde onu gösteriyor
```

```python
result[0][1]
```

# **Görselleştirme**

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(IMAGE_PATH)
img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB )
img = cv2.putText(img,text,top_left, font, 0.5,(255,255,255),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

```python
import Levenshtein

def easy_ocrAll(path,actual):
    IMAGE_PATH=path
    reader=easyocr.Reader(['en'],gpu=False)
    result=reader.readtext(IMAGE_PATH)
    pred=''
    for i in range(len(result)):
        read=result[i][1]
        pred=pred + read

    wer = Levenshtein.distance(pred, actual) / len(actual)

    
    return pred,wer*100,len(pred),len(actual)
        
        
        
        
        
```
